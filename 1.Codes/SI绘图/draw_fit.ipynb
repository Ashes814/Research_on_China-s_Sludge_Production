```python
import matplotlib as mpl
import seaborn as sns
import pandas
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = pd.ExcelFile('fit_results.xlsx')
fit_results = file.parse('Sheet1')
fit_results.head()
```

```python

mpl.rcParams['axes.linewidth'] = 0


mpl.rcParams['axes.axisbelow'] = False
mpl.rcParams['xtick.bottom'] = False
mpl.rcParams['ytick.left'] = False
mpl.rcParams['xtick.labelbottom'] = False
mpl.rcParams['ytick.labelleft'] = False
```

```python
def draw(df, var_T, var_P):
    plt.figure(figsize=(6,3))
    plt.plot(df.year, df[var_T],color='royalblue',linewidth=3)
    plt.plot(df.year, df[var_P],color='orange',linewidth=3, linestyle='--')
    plt.xticks(None)
    plt.yticks
    sns.despine()
    plt.savefig('{}.png'.format(var_P), dpi=300)
    
draw(fit_results, 'BA_T', 'BA_SSP1')
```
